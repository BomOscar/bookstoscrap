```python
from bs4 import BeautifulSoup as bs
import requests as rq

url = "https://books.toscrape.com/"

ree = rq.get(url)
rew = bs(ree.content, "html.parser")

docR = rew.find("section").find_all("div")
m = []
for i in docR:
    rr = i.find_all(class_="row")
    for i in rr:
        vv = i.find_all(class_="product_pod")
        for i in vv:
            ttl = i.find("h3").a['title'] #titles
           
            tt = i.find("h3").a['href'] #links
            
            pr = i.find(class_="product_price").p.text.replace("Â£", "")
            
            st = i.find(class_="instock availability").text.strip()
            
            mm = [ttl , tt , pr, st] #listOf
            m.append(mm)
```

```python
import pandas as pd
df = pd.DataFrame(m, columns = ["Title", "Link", "Price", "Stock"])
df.to_csv("BooksToScrap.csv")
df
```
